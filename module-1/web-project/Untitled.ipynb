```python
# <table cellpadding="0" cellspacing="0" style="font-family:Times New Roman;font-size:10pt;width:100%;border-collapse:collapse;text-align:left;">
# https://sec.report/Document/0000320193-18-000145/a10-k20189292018.htm

import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
from bs4 import BeautifulSoup
responseAAPL = requests.get('https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=AAPL&outputsize=full&apikey=XPYD1Q6WRNRWVYOV')
dataAAPL = responseAAPL.json();
forksAAPL=json_normalize(dataAAPL)
forksAAPL
```

```python

open=[]
high=[]
low=[]
close=[]
closeajusted=[]
volume=[]
dividend=[]
splitcoefficient=[]

for i in range(5, 43205, 8):
    open.append(forksAAPL.iloc[:,i][0])
    high.append(forksAAPL.iloc[:,i+1][0])
    low.append(forksAAPL.iloc[:,i+2][0])
    close.append(forksAAPL.iloc[:,i+3][0])
    closeajusted.append(forksAAPL.iloc[:,i+4][0])
    volume.append(forksAAPL.iloc[:,i+5][0])
    dividend.append(forksAAPL.iloc[:,i+6][0])
    splitcoefficient.append(forksAAPL.iloc[:,i+7][0])

time=[]
for i in range(5,43205,8):
    time.append(forksAAPL.columns[i])
    
dfAAPL = pd.DataFrame({'Open':open, 'High':high, 'Low':low, 'Close':close,'Close Ajusted':closeajusted, 'Volume':volume, 'Dividend':dividend, 'Split Coefficient':splitcoefficient})
dfAAPL.head()
```

```python
# <table cellpadding="0" cellspacing="0" style="font-family:Times New Roman;font-size:10pt;width:100%;border-collapse:collapse;text-align:left;">
# <div style="padding-left:0px;text-indent:0px;line-height:normal;padding-top:10px;">
# <tr class="Bdbw(1px) Bdbc($c-fuji-grey-c) Bdbs(s) H(36px)">
url = 'https://finance.yahoo.com/quote/AAPL/financials?p=AAPL'
html = requests.get(url).content
# soup = BeautifulSoup(html, "lxml")
soup = BeautifulSoup(html, "lxml")

table = soup.find_all('tr',{'Bdbw(1px) Bdbc($c-fuji-grey-c) Bdbs(s) H(36px)'})
table2=[]
for line in table:
    cleanline = (line.find_all('span'))
    table2.append([row.text.strip().split("\n") for row in cleanline])
table2
```

```python
info=[]
year18=[]
year17=[]
year16=[]
year15=[]

for i in table2:
    if len(i)==5:
        info.append(i[0])
        year18.append(i[1])
        year17.append(i[2])
        year16.append(i[3])
        year15.append(i[4])
```
